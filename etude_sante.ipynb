```python
import pandas as pa
import numpy as np

aide_alimentaire = pa.read_csv("data/aide_alimentaire.csv")
dispo_alimentaire = pa.read_csv("data/dispo_alimentaire.csv")
population = pa.read_csv("data/population.csv")
sous_nutrition = pa.read_csv("data/sous_nutrition_v1.csv")


```

```python
#remplacer les NaN par 0
dispo_alimentaire.fillna(value=0,inplace=True)
aide_alimentaire.fillna(value=0,inplace=True)
population.fillna(value=0,inplace=True)
sous_nutrition.fillna(value=0,inplace=True)
```

```python
#renommer les colonnes
aide_alimentaire.rename(columns={"Valeur":"V_aide_alimentaire","Pays bénéficiaire":"Pays"}, inplace=True)
dispo_alimentaire.rename(columns={"Zone":"Pays"}, inplace=True)
population.rename(columns={"Valeur":"V_population","Zone":"Pays"}, inplace=True)
sous_nutrition.rename(columns={"Valeur":"V_sous_nutrition","Zone":"Pays"}, inplace=True)
```

```python
def homogene(data_frame, colonnes, coef):
    for col in colonnes:
        data_frame[col]=(data_frame[col].astype('float'))*coef
    return data_frame

#multiplier par 10**6 toutes les colonnes dont l'unité est milliers de tonnes, pour homogénéiser en kg
liste_dispo_alimentaire=[
    "Aliments pour animaux",
    "Disponibilité intérieure",
    "Exportations - Quantité",
    "Importations - Quantité",
    "Nourriture",
    "Pertes",
    "Production",
    "Semences",
    "Variation de stock",
    "Autres Utilisations",
    "Traitement"
]

homogene(dispo_alimentaire,liste_dispo_alimentaire,10**6)


dispo_alimentaire.head()
```

```python
#multiplier par 1000 pour avoir en kg
homogene(aide_alimentaire,["V_aide_alimentaire"],10**3)
aide_alimentaire.head()
```

```python
#La population étant en milliers d'habitant,multiplier par 1 000  pour avoir en habitants
homogene(population, ["V_population"], 10**3)
population.head()
```

```python
#multiplier par 1 000 000 pour avoir en habitants
homogene(sous_nutrition, ["V_sous_nutrition"], 10**6)

sous_nutrition.head()
```

**Proportion de personnes en sous nutrition pour l'année 2017**
<br><li>une jointure entre population et sous nutrition</li><br>



```python
pop=population.loc[population['Année'] == 2017]
sn=sous_nutrition.loc[sous_nutrition['Année'] == '2016-2018']
pop_sn=pa.merge(pop, sn, on='Pays')
pop_sn.head()
```

```python
print("la Proportion de personnes en sous nutrition pour l'année 2017 est:",
      pop_sn['V_sous_nutrition'].sum()*100/pop_sn['V_population'].sum(),"%")
```

**le nombre théorique de personnes qui pourraient être nourries. Tu devrais pouvoir calculer ça à partir de la disponibilité alimentaire mondiale**
<li>convertir la colonne de disponibilité alimentaire en kcal. Celle ci est actuellement en kcal/pers/j</li>
<li>multiplier par la population et par 365.</li>
<li>Devise par (2500*365)car l'Humain a besoin presque 2500 kcal/j.</li>

```python
dispo_alimentaire_pop=pa.merge(dispo_alimentaire, pop, on='Pays')
dispo_kcal=dispo_alimentaire_pop['Disponibilité alimentaire (Kcal/personne/jour)']*dispo_alimentaire_pop['V_population']*365
dispo_kcal_tot= dispo_kcal.sum()
print("dispo alimentaire totale en kcal est:", dispo_kcal_tot,"kcal")
```

```python
dispo_kcal_hum=round(dispo_kcal_tot/(2500*365),2)
print("Total d'être humain pouvant être nourris est:",dispo_kcal_hum )
```

```python
prop=round(dispo_kcal_hum*100/pop['V_population'].sum())
print("proportion total d'être humain pouvant être nourris est:",prop,"%")
```

**L’utilisation de la disponibilité intérieure, en particulier la part qui est attribuée à l’alimentation animale, celle qui est perdue et celle qui est concrètement utilisée pour l'alimentation humaine.**
<br><li>créer une fonction définit proportion avec les paramétres(data_frame,list_disp_int,dispo_int)et on prend comme liste['Aliments pour animaux','Nourriture','Pertes']</li><br>

```python
dispo_int = dispo_alimentaire['Disponibilité intérieure'].sum()
list_disp_int=['Aliments pour animaux','Nourriture','Pertes']
def proportion(data_frame,list_disp_int,dispo_int):
     return data_frame[list_disp_int].sum()*100/dispo_int
```

```python
print("la proportion de Aliments pour animaux est:",proportion(dispo_alimentaire,'Aliments pour animaux',dispo_int),"%")
print("la proportion de Nourriture est:",proportion(dispo_alimentaire,'Nourriture',dispo_int),"%")
print("la proportion de Pertes est:",proportion(dispo_alimentaire,'Pertes',dispo_int),"%")
```
